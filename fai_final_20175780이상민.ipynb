```python
tStartDay=pd.to_datetime('2010-1-1', format='%Y-%m-%d')
tEndDay=pd.to_datetime('2021-11-30', format='%Y-%m-%d')
```

```python
momentumList=['best_3','best_6','best_9','best_12','median_3','median_6','median_9','median_12',
              'worst_3','worst_6','worst_9','worst_12']
myDf=pd.DataFrame(0,columns=momentumList,index=month_stocks.index)
def saveMeanEarningRate_best20(momentumDF,columname,idx):

  top20=momentumDF.loc[idx].sort_values(ascending = False).index[20:40]
    
  sum = 0.0
  for tStock in top20:
    value = momentumDF.loc[idx:].iloc[1].loc[tStock]
    if np.isnan(value) == False:
        sum += value
        
  mean= sum/20 
  myDf.loc[idx,columname]=mean

for idx, row in month_stocks.iterrows():
    if idx>=tStartDay and idx<tEndDay:
     
      saveMeanEarningRate_best20(momentum_3,momentumList[0],idx)
      saveMeanEarningRate_best20(momentum_6,momentumList[1],idx)
      saveMeanEarningRate_best20(momentum_9,momentumList[2],idx)
      saveMeanEarningRate_best20(momentum_12,momentumList[3],idx)
```

```python
def saveMeanEarningRate_median20(momentumDF,columname,idx):
    
  median = len(momentumDF) // 2
  median20=momentumDF.loc[idx].sort_values(ascending=False).index[median-10:median+10] 
  sum = 0.0
  for tStock in median20:
    value = momentumDF.loc[idx:].iloc[1].loc[tStock]
    if np.isnan(value) == False:
        sum += momentumDF.loc[idx:].iloc[1].loc[tStock] 
  
  mean= sum/20 
  myDf.loc[idx,columname]=mean

for idx, row in month_stocks.iterrows():
    if idx>=tStartDay and idx<tEndDay:
     
      saveMeanEarningRate_median20(momentum_3,momentumList[4],idx)
      saveMeanEarningRate_median20(momentum_6,momentumList[5],idx)
      saveMeanEarningRate_median20(momentum_9,momentumList[6],idx)
      saveMeanEarningRate_median20(momentum_12,momentumList[7],idx)
```

```python
def saveMeanEarningRate_worst20(momentumDF,columname,idx):

  worst20=momentumDF.loc[idx].sort_values(ascending = True).index[30:50] 

  sum = 0.0
  for tStock in worst20:
    value = momentumDF.loc[idx:].iloc[1].loc[tStock]
    if np.isnan(value) == False:
        sum += value

  mean= sum/20.0
  myDf.loc[idx,columname]=mean

for idx, row in month_stocks.iterrows():
    if idx>=tStartDay and idx<tEndDay:
     
      saveMeanEarningRate_worst20(momentum_3,momentumList[8],idx)
      saveMeanEarningRate_worst20(momentum_6,momentumList[9],idx)
      saveMeanEarningRate_worst20(momentum_9,momentumList[10],idx)
      saveMeanEarningRate_worst20(momentum_12,momentumList[11],idx)
```

```python
def saveMeanEarningRate_best20(myDf,momentumDF,columname,idx):
    top20=momentumDF.loc[idx].sort_values(ascending=False).index[20:40]
    sum = 0.0
    for tStock in top20:
        value = momentumDF.loc[idx:].iloc[1].loc[tStock]
        if np.isnan(value) == False:
            sum += value

        mean= sum/20 
        myDf.loc[idx,columname]=mean
    
    return myDf

def saveMeanEarningRate_median20(myDf, momentumDF,columname,idx):
    
    median = len(momentumDF) // 2
    median20=momentumDF.loc[idx].sort_values(ascending=False).index[median-10:median+10] 
    sum = 0.0
    for tStock in median20:
        value = momentumDF.loc[idx:].iloc[1].loc[tStock]
        if np.isnan(value) == False:
            sum += value

    mean= sum/20 
    myDf.loc[idx,columname]=mean
        
    return myDf

def saveMeanEarningRate_worst20(myDf, momentumDF,columname,idx):

    worst20=momentumDF.loc[idx].sort_values(ascending=True).index[30:50] 

    sum = 0.0
    for tStock in worst20:
        value = momentumDF.loc[idx:].iloc[1].loc[tStock]
        if np.isnan(value) == False:
            sum += value

    mean= sum/20.0
    myDf.loc[idx,columname]=mean
    
    return myDf
```

```python
def get_portfolio(momentum):
    myDf = pd.DataFrame(0,columns = ['best', 'median', 'worst'],index = month_stocks.index)
    
    for idx, row in month_stocks.iterrows():
        if idx >= tStartDay and idx < tEndDay:
            myDf = saveMeanEarningRate_best20(myDf, momentum, 'best' ,idx)
            myDf = saveMeanEarningRate_median20(myDf, momentum, 'median',idx)
            myDf = saveMeanEarningRate_worst20(myDf, momentum, 'worst',idx)
            
    return myDf
```

```python
def plot_cum_return_all(myDf, i, figsize=(8,4)):
    k1 = pd.merge(KOSPI_return, myDf, left_index=True, right_index=True, how = 'inner').dropna()
    
    k1_cumr = (1+k1/100).cumprod()

    k1_cumr.plot(figsize=figsize) 
    plt.title('KOSPI vs K='+str(i)+' : Cumulative Return') 
    plt.xlabel('Date') 
    plt.ylabel('누적수익') 
    plt.legend();
```
